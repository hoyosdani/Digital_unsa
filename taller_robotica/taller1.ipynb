## Programación OTA

Los microbots que se utilizaran deben poder desplazarse y por lo tanto grabarlos utilizando un cable resulta muy dificil. Se utilizara una tecnica denominada OTA (over the air). Donde los mirocontorladores tendran cargado un programa que los permita grabar a traves de intenet.

En la siguiente ventana se muestra como seria el formato inicial de un prograna. 


````python
%%writefile /home/dhoyos/jupyter_directorio/talleres_distancia/taller1/taller1.ino

#include "config.h"
void setup()
{
 configurarOTA();     
}

void loop()
{
ArduinoOTA.handle();
   
}
````

Un programa arduino tiene dos funciones principales:
- **setup():** Donde se configura el dispositivo.
-  **loop():** Donde se ejecutan las acciones.  
  Ademas se agrega un achivo de configuración

-  **config.h:** Se encuentran las definiciones.

Las funciones  **configurarOTA();**    y **ArduinoOTA.handle();** se utilizan para poder programa independientemente de un cable de conexion.

El microcontrolador que se usara en este ejemplo es el ESP8266 y la placa de desarrollo que se usara es NODEMCU que se muestra en la figura de abajo
<img src="img/nodemcu-pinout.webp" alt="Imagen"></img>  
La placa NODEMCU tiene un led integrado a la misma y se denomina BUILTIN_LED. El cual se usara en este ejemplo.
El entorno ARDUINO se programa en C++ y las sentencias de programación que se usaran se iran explicando durante el curso. En este ejemplo se usaran las siguientes sentencias


- **pinMode(pin, mode):** Se usa en la funcion void setup() para configurar un puerto específico para que se comporte o como entrada (INPUT) o como salida (OUTPUT).  
Ejemplo:  
 ```pintMode(pin, OUTPUT);``` Se programa el puerto como salida   
 ``` pinMode(pin, INPUT);``` Se programa el puerto como entrada   
- **digitalWrite(pin, valor):** Devuelveuna tension de 3.3V (HIGH) o 0 Voltios (LOW) en un pin digital especificado. 
``` digitalWrite(pin, HIGH); //ajusta 'pin' a HIGH ``` 
- **delay(ms):** Detiene el programa por la cantidad de tiempo especicada en milisegundos, donde 1000 es igual a 1 segundo.
Existen un conjunto de palabras reservadas por el lenguaje en este ejemplo son_
- **HIGH:** Una salida queda a 3,3 Volts
- **LOW:** Una salida queda a 0vVolts
- **OUTPUT:** Se configura un puerto como salida.
El programa quedaria:

El programa que estuvimos utilizando se denomina C++ y se lo denomina de alto nivel, dado que es muy cercano a la abstraccion que puede realizar nuestro cerebro. Previa a estas acciones debemos conocer como se denomina nuestro microcontrolador y donde grabarlo.

## Compilacion del microcontrolador

Para compilar el programa se usara la siguiente linea  
**!arduino-cli compile --fqbn  esp8266:esp8266:d1 --output-dir /home/dhoyos/jupyter_directorio/talleres_distancia/taller1/burn  /home/dhoyos/jupyter_directorio/talleres_distancia/taller1/taller1.ino**

donde **compile** indica que arduino-cli debe compilar un archivo  la opcion:  

**--fqbn** es la sigla de Fully Qualified Board Name (Nombre de Placa Totalmente Cualificado). En arduino-cli, es un identificador único y preciso que le indica al sistema exactamente qué modelo de placa, arquitectura y configuración  está utilizando

**--output-dir:** El directorio donde se grabara el archivo.  
Finalmente se coloca el nombre del archivo a compilar.

Si el codigo tiene un error el resultado sera la linea donde se detecto el error si la compilacion resulta correcta se muestra un mensaje limpio en la terminal, con un resumen del uso de memoria del dispositivo


````python
!arduino-cli compile --fqbn  esp8266:esp8266:d1 --output-dir /home/dhoyos/jupyter_directorio/taller1/burn  /home/dhoyos/jupyter_directorio/taller1/taller1.ino
````
