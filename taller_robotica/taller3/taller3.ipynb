## Talller servomotor RC

Un servomotor RC (o servo) es un motor eléctrico con una reducción que incorpora un sistema de control y un sensor de posición que permite controlar con máxima precisión la posición angular.  

<img src="img/Partes-servomotor.webp" alt="Imagen"></img>
Partes de un servomotor RC:

- Circuito de control: Es un circuito diseñado con una retroalimentación para poder controlar el giro deseado del servomotor.   

- Motor de corriente continua: Prácticamente es el corazón del servomotor ya que este es el que nos proporciona el movimiento.

- Engranajes: Se utilizan engranes reductores para poder reducir la velocidad de giro del motor de corriente continua, con la finalidad de ganar mayor fuerza (torque).

- Sensor de desplazamiento: En este caso se utiliza un potenciómetro que se conecta al eje del motor para que gire junto con el y el circuito de control pueda medir su cambio de resistencia y calcular el giro


Los servomotores RC funcionan a través de una señal eléctrica que se debe modificar según la posición en la que debe girar el motor. Los servomotores funcionan usando PWM (Pulse Wide Modulation) o Modulación por ancho de pulso. 


En donde el periodo de la señal debe ser de 20 ms, 
|Angulo | Ancho del pulso |
|---    |---              |
| 0°    | 0,9 ms          |
| 90°   | 1,5 ms          |
| 180°  | 2,5 ms          |

|Negro   | Masa            |

Funcionamiento servomotor RC:

- Recepción: El sistema recibe una señal eléctrica  que define la posición deseada.

- Accionamiento: El motor eléctrico interno gira y un sistema de engranajes reduce la velocidad.

- Retroalimentación: El sensor interno lee la posición real y el controlador corrige cualquier margen de error al instante.




## Programación de servomotores RC

Para programar servomotores RC con una placa nodemcu usando **arduino** se debe utilizar la libreria servo.h. La cual fue diseñada para controlarlos.

Se comienza cargando la libreria servo con la sentencia **#include <Servo.h>**

Luego se crea el objeto servo a partir del cual se utiliza el servomotor por el microcontrolador la sentencia es **Servo nombre_servo**

Para informar a que pin del micrcontrolador se conecta se utiliza la sentencia attach donde debe escribirse el múmero de pin donde se conectara el servomotor  **nombre_servo.attach(numero)**

Utilizando la sentencia write se ordena al servo el angulo en el cual se debe ubicar, la sentencia se muestra a continuacion
**nombre_servo.write(90)**

Finalmente la sentencia detach desconecta el servomotor **brazo.detach()**
Se debe declarar el servomotor que se usara.
Las consideraciones a tener en cuenta antes de conectar un servomotor son:    
 - Voltaje: Los servos estándar (como el SG90) operan entre 4.8V y 6V.   
- Conexión a tierra (GND): Es obligatorio conectar el pin GND de la fuente externa de 5V al pin GND del NodeMCU
- Pin de datos: Se puede utilizar cualquier pin GPIO compatible con PWM en el ESP8266: D1, D2, D3 o D4

Se muestra el codigo fuente de un ejemplo de programa que controla un servomotor RC
