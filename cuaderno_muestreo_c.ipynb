## Definiciones importantes en instrumentación
1- El error de una medicion se define como:

$$ \varepsilon =xi-yi $$

Yi=valor  verdadero definido o calculado

Xi=valor actual medido

2- Exactitud (Accuracy):

La exactitud de una medida se define como :
           
$$ A=1- \frac{yi-xi}{yi} $$

3- Precisión y media (Precision and Mean)

El valor medio se define:           

$$ xm= \frac {1} {n} \sum_{i=1}^{i=n} x_{i} i=1:n $$

n = numero de medidas      xi medida iesim

La Precisión se define como:   
       
$$ Pi= 1- \frac{x-xm}{xm} $$

4- Resolución (Resolution)
       Se define como la menor cantidad que el sistema puede detectar en la medición

$$ V_{min}= \frac{ V_{fesc}}{(2^{n}-1}) $$
       
5- Varianza 

$$ \sigma^{2}= \frac{1}{N} \sum _{n=1}^{n=N}(x_{n}-X_{m}) $$

6- Desviacion Standart (Variance and Standard deviation)

$$ \sigma^{2}=\sqrt { \frac{1}{N} \sum _{n=1}^{n=N}(x_{n}-x_{m})} $$
 



       
       
7- Finura  (Finesse)
       El grado que una cantidad que se esta siendo medida es modificada por el proceso de medición
       
8- Sensitividad (Sensitivity)
       Se define como la minima entrada que provoca un cambio en la salida detectable. 
       
9- Rango (Range)
       Se define como rango del sensor al maximo y minimo valor del parametro que puede ser leido.

10- Offset y factor de escala (bias and scale factor shift)
       
El offset se define como el valor de salida que se mide cuando la entrada es cero.
       
11- Linealidad y Regresion lineal (Linearity and Linear Regression)

12- Histeresis (Hysteresis)

## Ruido
Se denomina ruido a una señal no deseada que se suma a la señal que se desea medir. Existen diversas fuentes de ruido:

### Ruido blanco

Se caracteriza por tener la misma potencia de señal en todas las frecuencias. Tambien se denomina ruido termico y se debe a la agitación termica de los atomos.

Lo que significa que si se eliminan las frecuencias mayores que las significativas de la señal se disminuye el ruido.
En la figura se representa una señal de ruido.

```python
def ruido_blanco(mrb):
    random.seed(19680801)
    t1 = arange(0.0, 200.0, 1.0)
    vrb = mrb*random.rand(*t1.shape) 
    plot(t1, vrb)
```

```python
interact(ruido_blanco, mrb=(0, 100));
```

```python
ruido_blanco(100)
```

## MUESTRO Y RETENCION
Es el proceso mediante el cual se corta la señal en el tiempo para que se pueda medir. En esta seccion se analiza las consecuencias de este proceso. Con este fin generamos una señal como la mostrada en la figura.

```python
t=arange(200)
w=2*pi/100
v=100*sin(w*t)
plot(t,v)
```

la señal analogico para ser tranformada debe ser muestreada en el tiempo. En la figura se muestra una señal muestreada en el tiempo

```python
def fun_muestreo(nsamples,tmuestreo):
    w=2*pi/100
    t=arange(nsamples)
    v=100*sin(w*t)
    v1=zeros(nsamples)
    v1[:nsamples:tmuestreo]=v[:nsamples:tmuestreo]
    plot(t,v1)
    plot(t,v)
```

```python
fun_muestreo(400,10)
```

En naranja se observa la señal. En azul la señal muestreda. Si el tiempo de muestreo es mucho menor a la mitad del periodo, en este caso el periodo es 100 y el tiempo de muestreo es 10. Se puede observar que la señal se reproduce correctamente. Si el tmuestreo es del orden de la mitad del periodo, el resultado se muestra en la siguiente figura

```python
fun_muestreo(400,50)
```

El resultado es una linea recta que no tiene nada que ver con la señal. Si el tmuestreo es mayor que la mitad del periodo ocurre lo que muestra la figura

```python
interact(fun_muestreo, nsamples=(0, 1000),tmuestreo=(0,200));
```

## Filtros pasabajos
Un filtro pasabajo es un dispositivo que permite pasar a las freecuencias menores a la frecuencia de corte. Una caracteristica ideal se muestra en la figura

```python
v=concatenate((ones(100),zeros(100)), axis=0)
t=arange(200)
plot(t,v)
xlabel("Frecuencia")
ylabel("Señal")
```

Esta curva se denomina plantilla, los filtros reales no tienen esa forma sino la que se muestra en la figura dependiendo de la cantidad de polos de la ecuacion. El circuito mas sencillo es el RC. Que se muestra en la figura. 


La funcion de transferencia, es la funcion que caracteriza el circuito y es la relacion entre la salida y la entrada es:

$$ FT= \frac{1}{1+jw/w0} $$

Donde 
$$  w0= \frac{1}{R.C}   $$  
R: resistencia del circuito.

C: Capacidad del circuito.

w0: es la frecuencia de corte

$$ w: 2 \pi f $$

La frecuencia de corte esta en



```python
def FT(w,r,c):
    wc = 2*pi*1000/r/c
    return 1.0 / (1.0 + 1j * w / wc)
```

```python
def iFT(r,c):
    f = logspace(1,5) # frequencies from 10**1 to 10**
    plot(f, 20*log10(abs(FT(f,r,c)))); xscale('log')
    xlabel("log Frecuencia")
    ylabel("Funcion Tranferencia")
    print ("Frecuencia de corte")
```

```python
iFT(10,10)
```

```python
interact(iFT,r=(0, 20),c=(0,20));
```

## CONVERSOR DA  

Convierte una señal digital en una señal analogica (un conversor AD tiene un conversor DA entre sus partes. El conversor digital analogico o DA convierte un codigo binario a un valor de tension. En la figura se muestra un esquema donde se pueden observar, las entradas digitales, la fuente de refencia y la salida.


El numero de entradas digitales determina el maximo valor que puede tener el conversor. En este caso cuatro entradas determina que el codigo sea de cuatro bits. O sea que puede representar un numero entre 0 y 15. El maximo valor lo determina la tension de referencia. Segun la siguiente ecuacion:

$$ Vs=\frac {Vref}{2^n-1}.valor   $$
Donde:

    Vref: tensión de referencia
    n: número de bits
    valor: valor binario que se desea representar
 
    
La curva de respuesta se muestra abajo

```python
def cda(nbits,vref):
    t1=arange(0,255)
    s=arange(0,255)
    nn=256/(2**nbits)
    for t in range (255):
        s[t]=vref/(2**nbits)*int(t/nn)

    plot(s)
    xlabel("Codigo Digital")
    ylabel("Salida")
```

```python
interact(cda,nbits=(1,12),vref=(1,100));
```

```python
cda(4,100)
```

```python
def seno_da(nbits):
    s =arange(200)
    vref=10
    w=2*pi/100
    nn=256/(2**nbits)
    for t in range (200):
        s[t]=nn*int((128*sin(w*t)+128)/nn)
    plot(s)
            
```

 ### Errores de los conversores DA
 
 Los errores de los dispositivos se deben a pequeñas variaciones en los componentes del sistema, alinealidades, pequeñas variaciones debida a la temperatura. Todas estas se agrupan de forma de poder medirlas y tenerla en cuenta para el diseño y uso del dispositivo
 
 1. Estáticos
     1. Lineales
         1. Offset: Se mide ingresando cero a las entrads digitales
         2. Ganancia: Se mide colocando el maximo valor en las entradas digitales, comparando este valor con el valor maximo deseado
     2. No lineales
         1. Error integral no lineal (INL): Maxima separacion entre la curva ideal y la real
         2. Error diferencial no lineal (DNL):máxima diferencia entre el cambio de VO en un cambio de 1bit y el cambio esperado
 2. Dinámicos
     1. Tiempo de establecimiento: Tiempo que tarda la salida en llegar a ±0.5LSB del valor final al cambiar la entrada 

Los errores mas significativos que presenta los Conversores Digital Analogico son offset y ganancia.En la figura se representa como varia la curva y la respuesta del sistema a la variacion del offset y la ganacia en este caso offset=10% y ganancia=10% (son valores altos con respecto a los reales con el solo efecto de representarlo en el grafico). Estos errores se pueden recalibrar.

```python
def cda_e(nbits,vref, offset,gain):
    t1=arange(0,255)
    s=arange(0,255)
    s1=arange(0,255)
    s2=arange(0,400)
    s3=arange(0,400)
    nn=256/(2**nbits)
    for t in range (255):
        s[t]=vref/(2**nbits)*int(t/nn)
        s1[t]=vref*(1+gain/100)/(2**nbits)*int(t/nn)+vref*offset/100
    subplot(1,2,1)
    plot(s)
    plot(s1) 
    xlabel("Codigo Digital")
    ylabel("Salida")
    w=2*pi/100
    for t in range (400):
        s2[t]=vref/255*nn*int((128*sin(w*t)+128)/nn)
        s3[t]=vref*(1+gain/100)/255*nn*int((128*sin(w*t)+128)/nn)+vref*offset/100
    subplot(1,2,2)
    plot(s2)
    plot(s3)
    xlabel("Tiempo")
    ylabel("Salida")
```

4. sigma delta: Este conversor es el mas utilizado en la actualidad cuando se requieren altas resoluciones. Se basa en sobremuetrear la señal parea disminuir el ruido. Consta de un filtropasabajo, integrador, comparador, conversor DA  de un bit y Decimator. Toma la señal y la compara con la salida de un integrador la misma es enviada al conversor AD si es 1 el ad se conecta a la referencia positiva y si es cero se conecta a la referencia negativa. El integrador "integra" la señal y la aproxima a la señal a medir. La salida del comparador se la envia a un contador que cuando la señal del compardor es 1 suma 1 al contador en cambio si es 0 resta 1 al comparador. 



Se presenta un modulador sigma delta


```python
def sigma_delta(muestreo,valor):
    t=arange(0,1000)
    s=arange(0,1000)
    s1=zeros(1000)
    integral=0
    w=2*pi/400
    s[t]=50*sin(w*t)
    for t1 in range(0,1000,muestreo):
        if s[t1]> integral:
            integral=integral+valor
        else:
            integral=integral-valor
        s1[t1]=integral
    plot(t,s1)
    plot(t,s)
```

### Errores de los conversores AD

Los errores de los dispositivos se deben a pequeñas variaciones en los componentes del sistema, alinealidades, pequeñas variaciones debida a la temperatura. Todas estas se agrupan de forma de poder medirlas y tenerla en cuenta para el diseño y uso del dispositivo. Se pueden clasificar como errores:

1. Estáticos
    1. Lineales
        1. Offset: Se mide ingresando cero a las entrads digitales
        2. Ganancia: Se mide colocando el maximo valor en las entradas digitales, comparando este valor con el valor maximo deseado
    2. No lineales
        1. Error integral no lineal (INL): Maxima separacion entre la curva ideal y la real
        2.  Error diferencial no lineal (DNL):máxima diferencia entre el cambio de VO en un cambio de 1bit y el cambio esperado
